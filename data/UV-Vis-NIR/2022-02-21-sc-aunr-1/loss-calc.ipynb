```python
# Import all the necessary functions

import glob
import os, time
import itertools
import numpy as np
import seaborn as sns
import pandas as pd

import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import matplotlib.cm as cm
from mpl_toolkits.mplot3d import Axes3D
from mpl_toolkits.axes_grid1.inset_locator import inset_axes
from cycler import cycler
from datetime import datetime, timedelta

from scipy.stats import norm
from scipy.optimize import curve_fit #we could import more, but this is what we need
from scipy.stats import linregress

%matplotlib inline
```

```python
df_obj = pd.read_csv('rev-obj.csv', header=None)
df_1 = pd.read_csv('exp1.csv', header=None)
```

```python
df_obj
```

# Experiment 1

```python
y_exp1 = df_1[1]
norm_y_exp1 = (y_exp1 - y_exp1.min())/(y_exp1.max()-y_exp1.min())

max_norm_dist_1 = np.linalg.norm(df_obj[1] - norm_y_exp1)
max_norm_dist_1 = round(max_norm_dist_1, 2)
print('loss: ',max_norm_dist_1)

plt.plot(df_obj[0], norm_y_exp1, 'r', label = 'Experiment')
plt.plot(df_obj[0], df_obj[1], 'k', label = 'Objective')
plt.xlim(400,1100)
plt.ylim(0,1.05)
plt.xlabel('Wavelength (nm)')
plt.ylabel('Norm. Extinction')
plt.legend()
```
